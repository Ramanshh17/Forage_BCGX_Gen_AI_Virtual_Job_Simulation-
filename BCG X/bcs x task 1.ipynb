```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_excel(r'/content/Financial_Analysis_Microsoft_Tesla_Apple_2022-2024.xlsx')
```

```python
data
```

```python
data['Revenue Growth (%)'] = data.groupby('Company')['Total Revenue ($B)'].pct_change() * 100
data['Net Income Growth (%)'] = data.groupby('Company')['Net Income ($B)'].pct_change() * 100
```

```python
data
```

```python
data.fillna(0, inplace=True)
```

```python
data
```

```python
data['Assets Growth (%)'] = data.groupby('Company')['Total Assets ($B)'].pct_change() * 100
data['Liabilities Growth (%)'] = data.groupby('Company')['Total Liabilities ($B)'].pct_change() * 100
data['Cash Flow from Operations Growth(%)'] = data.groupby('Company')['Operating Cash Flow ($B)'].pct_change() * 100
```

```python
data
```

```python
data.fillna(0, inplace=True)
```

```python
data
```

```python
data.to_csv('/content/Financial_Analysis_Microsoft_Tesla_Apple_2022-2024.xlsx')
```

```python
summary = data.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Assets Growth (%)' : 'mean',
    'Liabilities Growth (%)' : 'mean',
    'Cash Flow from Operations Growth(%)' :'mean'
}).reset_index()
```

```python
print('Year-By-Year Average Growth Rates(%) :-')
print('             OR')
print("Overall Growth/Fall rate for Apple, Microsoft and Tesla from Fiscal Year 2021 - 2023")
summary
```
